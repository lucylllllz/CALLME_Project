```python
import pandas as pd
import matplotlib.pyplot as plt

CSV1_PATH = "outputs/generate_all_story_results.csv" ## Give everything but hold out teacher
CSV2_PATH = "outputs/no_imgs_story_results.csv"## Give no imgs and hold out teacher
CSV3_PATH = "outputs/no_bbox_story_results.csv" ## Give 1 imgs and hold out teacher

# ========= LOAD DATA =========
df = pd.read_csv(CSV1_PATH)
df1 = pd.read_csv(CSV2_PATH)
df2 = pd.read_csv(CSV3_PATH)

df
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
df["perplexity"].plot(kind="hist", bins=20, title="perplexity generate_all",ax=axes[0])
df1["perplexity"].plot(kind="hist", bins=20, title="perplexity no_imgs",ax=axes[1])
df2["perplexity"].plot(kind="hist", bins=20, title="perplexity no_bbox",ax=axes[2])
```
